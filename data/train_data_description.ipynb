Train.csv has 53 columns in total, and 401,125 rows.

The value to predict is the column "SalePrice".

**Important!** : **We use the log to get the % of error as opposed to absolute error**

```
import pandas as pd

df_raw = pd.read_csv(f'Train.csv', low_memory=False, 
                     parse_dates=["saledate"])
df_raw.shape
```

```
#Root mean squared log error
#We use the log to get the % of error as opposed to absolute error which doesn't make much sense
import numpy as np
df_raw.SalePrice = np.log(df_raw.SalePrice)
```
